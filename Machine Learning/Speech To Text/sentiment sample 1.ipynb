```python
from pydub import AudioSegment
import shutil
import wave
import os
import contextlib as cl
from pydub.silence import split_on_silence
```

```python
def segmentation(min_silence_len, silence_thresh, chunk_silent):
    song = AudioSegment.from_file('test212.wav', format="wav")
    shutil.rmtree("audio_chunks")
    chunk_dur = []
    
    chunks, silent_ranges = split_on_silence(song, min_silence_len=min_silence_len, silence_thresh=silence_thresh)
    
    try:
        os.mkdir('audio_chunks')
    except FileExistsError:
        pass
    
    i = 0
    chunk_silent = AudioSegment.silent(duration=100)
    
    for chunk in chunks:
        audio_chunk = chunk_silent + chunk + chunk_silent 
        audio_chunk.export("./audio_chunks/chunk{0}.wav".format(i), bitrate='64k', format="wav")
        i += 1
    
    filename = './audio_chunks/chunk' + str(i-1) + '.wav'
    
    with cl.closing(wave.open(filename, 'r')) as f:
        rate1 = f.getframerate()
    chunk_rate = rate1
    nf_chunks = i
    
    return chunk_rate, nf_chunks, silent_ranges
```

```python
import nltk
nltk.download('vader_lexicon')  # Download the Vader lexicon for sentiment analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
data = []
with open('1.txt', 'r', encoding='utf-8') as file:
    for i in file:
        data.append(i.strip().split())
```

```python
data
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.stem import PorterStemmer
from nltk.tokenize import RegexpTokenizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('hinglish'))
```

```python
stop_words = list(stop_words)
```

```python
filtered_tokens = [word for word in data if word not in stop_words]
```

```python
print(type(stop_words))
print(type(data))
```

```python
print(filtered_tokens)
```

```python
filter_tokens = []
for i in filtered_tokens:
    filter_tokens.extend(i)
filter_tokens
```

```python
sid = SentimentIntensityAnalyzer()
sentiments = []
```

```python
for i in filter_tokens:
    # Analyze sentiment for each sentence
    sentiment_score = sid.polarity_scores(i)
    # Classify sentiment as positive, negative, or neutral based on compound score
    if sentiment_score['compound'] >= 0.05:
        sentiment = 'positive'
    elif sentiment_score['compound'] <= -0.05:
        sentiment = 'negative'
    else:
        sentiment = 'neutral'
    sentiments.append((i, sentiment))
```

```python
print(sentiments)
```
